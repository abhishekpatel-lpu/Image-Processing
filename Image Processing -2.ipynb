```python
#Translaton of the image

import cv2
import numpy as np

image1=cv2.imread("10.jpg")

# Translation with translation matrix 
row,col,ch=image1.shape
# x,y direction of shifting value
x=50
y=25

T=np.float64([[1,0,x],[0,1,y]])

# images source second Matrix, third sestination size, fourth
image2=cv2.warpAffine(image1,T,(col,row))



cv2.imshow("frame1",image1)
cv2.imshow("frame2",image2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Translaton of the image with live using loop condition

import cv2
import numpy as np

image1=cv2.imread("10.jpg")

# Translation with translation matrix 
row,col,ch=image1.shape
# x,y direction of shifting value
x=50
y=25



while(True):
    T=np.float64([[1,0,x],[0,1,y]])

    # images source second Matrix, third destination size, fourth
    image2=cv2.warpAffine(image1,T,(col,row))



    cv2.imshow("frame1",image1)
    cv2.imshow("frame2",image2)
    k=cv2.waitKey(1)
    if k==ord("q"):
        break
    x=x+.025
    y=y+.025  # we can hold any axis can can observe the motion
    
    
cv2.destroyAllWindows()
```

```python
# Rotation Of Image

import cv2
import numpy as np

image1=cv2.imread("10.jpg")

# Translation with translation matrix 
row,col,ch=image1.shape

# first center point to rotate, second angle, third scaling factor 1 (default)
R=cv2.getRotationMatrix2D((row/2,col/2),180,1)

# warpAffine method is Used
#images source second Matrix, third sestination size, fourth

image2=cv2.warpAffine(image1,R,(col,row))



cv2.imshow("frame1",image1)
cv2.imshow("frame2",image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotation Of Image with loop

import cv2
import numpy as np

image1=cv2.imread("10.jpg")

# Translation with translation matrix 
row,col,ch=image1.shape
angle =0

while(True):
    if angle==360:
        angle=0
        
    # first center point to rotate, second angle, third scaling factor 1 (default)
    R=cv2.getRotationMatrix2D((row/2,col/2),angle,0.8)
    
    #R=cv2.getRotationMatrix2D((row/2,col/4),angle,0.8)---- changing the centre point

    # warpAffine method is Used
    #images source second Matrix, third sestination size, fourth

    image2=cv2.warpAffine(image1,R,(col,row))



    cv2.imshow("frame1",image1)
    cv2.imshow("frame2",image2)
    k=cv2.waitKey(1)
    if k==ord("q"):
        break
    angle=angle+1


cv2.destroyAllWindows()
```

```python
# scaling of the Image

import cv2
import numpy as np

image1=cv2.imread("25.jpg")

# interpolation method four types
# (Inter_linear, Inter_nearest, Inter_area, Inter_cubic, Inter_lanczos4)------default, near value, shrinking/reducing, zooming

# first-source image, second-destination

image2=cv2.resize(image1,None, fx=1.5,fy=1.5,
                  interpolation=cv2.INTER_LINEAR)

image3=cv2.resize(image1,None, fx=1.5,fy=1.5,
                  interpolation=cv2.INTER_AREA)

image4=cv2.resize(image1,None, fx=1.5,fy=1.5,
                  interpolation=cv2.INTER_NEAREST)

image5=cv2.resize(image1,None, fx=1.5,fy=1.5,
                  interpolation=cv2.INTER_CUBIC)

image6=cv2.resize(image1,None, fx=1.5,fy=1.5,
                  interpolation=cv2.INTER_LANCZOS4)


#pyramid Method

image7=cv2.pyrUp(image1)  # double the size of image
cv2.imshow("pyrup",image7)

image8=cv2.pyrDown(image1)  # decrease the size of image
cv2.imshow("pyrdown",image8)


cv2.imshow("frame1",image1)
cv2.imshow("Linear",image2)
cv2.imshow("Area",image3)
cv2.imshow("Nearest",image4)
cv2.imshow("Cubic",image5)
cv2.imshow("Lanczos4",image6)
cv2.waitKey()
cv2.destroyAllWindows()
```
