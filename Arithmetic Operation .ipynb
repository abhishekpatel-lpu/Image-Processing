```python
# Bit Wise Operation in the image

import cv2
import numpy as np

image1=cv2.imread("1.jpg")

image2=np.zeros(image1.shape,np.uint8)
image3=np.zeros(image1.shape,np.uint8)

cv2.rectangle(image2,(50,50),(300,300),(255,255),-1)
cv2.circle(image3,(275,275),50,(255,255),-1)

# bitwise operation we need black and white images
image4=cv2.bitwise_and(image2,image3)
image5=cv2.bitwise_or(image2,image3)
image6=cv2.bitwise_xor(image2,image3)
image7=cv2.bitwise_not(image2)   #not operator onlyu take one argument



cv2.imshow("frame1",image2)
cv2.imshow("frame2",image3)
cv2.imshow("frame3",image4)
cv2.imshow("frame4",image1)
cv2.imshow("frame5",image5)
cv2.imshow("frame6",image6)
cv2.imshow("frame7",image7)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Bit Wise Operation in the image with masking

import cv2
import numpy as np

image1=cv2.imread("10.jpg")
image5=cv2.imread("11.jpg")

image2=np.zeros(image1.shape,np.uint8)
image3=np.zeros(image1.shape,np.uint8)

cv2.rectangle(image2,(50,50),(300,300),(255,255),-1)
cv2.circle(image3,(275,275),50,(255,255),-1)

# masking in bitwise operator

image4=cv2.bitwise_and(image1,image3)


cv2.imshow("frame1",image2)
cv2.imshow("frame2",image3)
cv2.imshow("frame3",image4)
cv2.imshow("frame4",image1)
cv2.imshow("frame5",image5)
cv2.waitKey()
cv2.destroyAllWindows()
```
