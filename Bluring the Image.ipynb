```python
# Blurring the Images

import cv2
import numpy as np

image1=cv2.imread("26.jpg")

kernel_5x5=(np.ones((5,5),np.uint8))/25
kernel_15x15=(np.ones((9,9),np.uint8))/81

image2=cv2.filter2D(image1,-1,kernel_5x5)
image3=cv2.filter2D(image1,-1,kernel_15x15)

image4=cv2.blur(image1,(5,5))
image5=cv2.GaussianBlur(image1,(5,5),0)
image6=cv2.medianBlur(image1,5)
image7=cv2.bilateralFilter(image1,9,75,75)

cv2.imshow("original",image1)
cv2.imshow("filter2D",image2)
cv2.imshow("blur",image4)
cv2.imshow("gaissian Blur",image5)
cv2.imshow("median blur",image6)
cv2.imshow("bilateralFilter",image7)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Sharpening of the Image

import cv2
import numpy as np

image1=cv2.imread("26.jpg")

kernel=np.array([[-1,-1,-1],
                 [-1,9,-1],
                 [-1,-1,-1]])
image2=cv2.filter2D(image1,-1,kernel)

cv2.imshow("original",image1)
cv2.imshow("filter",image2)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
# Thresholding in the Image
# mainly used for finding the hidden feature from the image

import cv2
import numpy as np

image1=cv2.imread("15.jpg")

image1=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)

#first source image, second threshold value, third maximum value, fouth type

ret,image2=cv2.threshold(image1,100,255, cv2.THRESH_BINARY)

ret,image3=cv2.threshold(image1,100,255, cv2.THRESH_TRUNC)

ret,image4=cv2.threshold(image1,100,255, cv2.THRESH_TOZERO)

ret,image5=cv2.threshold(image1,100,255, cv2.THRESH_TOZERO_INV)

ret,image6=cv2.threshold(image1,100,255, cv2.THRESH_BINARY_INV)



cv2.imshow("original",image1)
cv2.imshow("T-Binary",image2)
cv2.imshow("T-Trunc",image3)
cv2.imshow("T-TOZERO",image4)
cv2.imshow("T-TOZERO_INV",image5)
cv2.imshow("T-Binary_INV",image6)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
# Adaptive Thresholding 

# Thresholding in the Image
# mainly used for finding the hidden feature from the image

import cv2
import numpy as np

image1=cv2.imread("16.jpg")

image1=cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)


image2=cv2.adaptiveThreshold(image1,255,cv2.ADAPTIVE_THRESH_MEAN_C,
                            cv2.THRESH_BINARY,13,-2)

image3=cv2.adaptiveThreshold(image1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY,13,-2)


cv2.imshow("original",image1)
cv2.imshow("Adaptive threshold",image2)
cv2.imshow("Gaussion-T",image3)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Erosion and Dilation of the Image

# Erosion
import cv2
import numpy as np

image1=cv2.imread("17.jpg")
kernel=np.ones((5,5),np.uint8)

image2=cv2.erode(image1,kernel,iterations=1)
image3=cv2.dilate(image1,kernel,iterations=1)
image4=cv2.morphologyEx(image1,cv2.MORPH_OPEN,kernel)
image5=cv2.morphologyEx(image1,cv2.MORPH_CLOSE,kernel)


cv2.imshow("original",image1)
cv2.imshow("Erosion",image2)
cv2.imshow("Dilation",image3)
cv2.imshow("M-OPEN",image4)
cv2.imshow("M_CLOSE",image5)
cv2.waitKey()
cv2.destroyAllWindows()
```
